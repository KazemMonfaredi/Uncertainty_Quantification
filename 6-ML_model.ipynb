```python
# first neural network with keras tutorial
import numpy as np
import pandas as pd 
from keras.models import Sequential
from keras.layers import Dense
# load the dataset
dataset = pd.read_csv("Realization_Feat.csv", header=0 , index_col=0)
selected_realizations=pd.read_csv("Initial_Realization_HM.csv" , header=None)
Selected_Realizations_dataset=dataset.loc[list(selected_realizations[0])]
Selected_Realizations_dataset["history_matching_results"]=list(selected_realizations[1])

    
# define the keras model
    
# split into input (X) and output (y) variables
Import =Selected_Realizations_dataset[['Ave_Perm_KG_02', 'Ave_Perm_Phi_KG_02', 'ARMSE_KG_02',
                                  'ARMSE_Phi_KG_02', 'STD_Perm_KG_02', 'STD_Perm_Phi_KG_02',
                                  'RMSE_std_KG_02', 'RMSE_std_Phi_KG_02', 'Ave_Perm_KG_04',
                                  'Ave_Perm_Phi_KG_04', 'ARMSE_KG_04', 'ARMSE_Phi_KG_04',
                                  'STD_Perm_KG_04', 'STD_Perm_Phi_KG_04', 'RMSE_std_KG_04',
                                  'RMSE_std_Phi_KG_04', 'Ave_Perm_KG_50', 'Ave_Perm_Phi_KG_50',
                                  'ARMSE_KG_50', 'ARMSE_Phi_KG_50', 'STD_Perm_KG_50',
                                  'STD_Perm_Phi_KG_50', 'RMSE_std_KG_50', 'RMSE_std_Phi_KG_50',
                                  'Ave_Perm_KG_56', 'Ave_Perm_Phi_KG_56', 'ARMSE_KG_56',
                                  'ARMSE_Phi_KG_56', 'STD_Perm_KG_56', 'STD_Perm_Phi_KG_56',
                                  'RMSE_std_KG_56', 'RMSE_std_Phi_KG_56', 'Ave_Perm_KG_58',
                                  'Ave_Perm_Phi_KG_58', 'ARMSE_KG_58', 'ARMSE_Phi_KG_58',
                                  'STD_Perm_KG_58', 'STD_Perm_Phi_KG_58', 'RMSE_std_KG_58',
                                  'RMSE_std_Phi_KG_58', 'variog']]
Export = Selected_Realizations_dataset["history_matching_results"]


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Import, Export, test_size=0.33, random_state=1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(Import)
scaled = scaler.fit_transform(Import)


model = Sequential()
model.add(Dense(10, input_dim=41,activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
# compile the keras model

recall=0
precision=0
n=0
while (((recall<0.85) | (precision<0.8)) &(n<1000)):
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc',f1_m,precision_m, recall_m])
    #it the keras model on the dataset
    model.fit(X_train, y_train, epochs=150, batch_size=10)
    # evaluate the keras model
    loss, accuracy, f1_score, precision, recall = model.evaluate(Import, Export)
    acc=accuracy
    print('Accuracy: %.2f' % (accuracy*100))
    n+=1
    print(n)
print('Accuracy: %.2f' % (accuracy*100))
print('f1_score: %.2f' % (f1_score))
print('precision: %.2f' % (precision))
print('recall: %.2f' % (recall))
print(n)

Y=(model.predict(Import)>0.5).astype(int)
y_n=np.array(Export)
Yn=Y.reshape(60,)
mm=y_n-Yn
mn=abs(y_n-Yn)
import matplotlib.pyplot as plt
plt.scatter(range(0,60),mm)
#plt.scatter(range(0,401),Yn ,marker='*')
print(sum(mm))
print(sum(mn))
print((sum(mm)+sum(mn))/2)
```
