```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn

FPR_hm=pd.read_table("FPR_KG.csv" , header=None , delimiter=", ")
FOPT_hm=pd.read_table("FOPT_KG.csv" , header=None , delimiter=", ")
FGPT_hm=pd.read_table("FGPT_KG.csv" , header=None , delimiter=", ")
Days=list(pd.read_csv("Days", header=None)[0])
```

```python
from matplotlib import cbook
import matplotlib.pyplot as plt
import numpy as np



fig, ax = plt.subplots(figsize=[8, 8])
ax.set_ylabel('FGPT (MMMscf)', fontsize=18 , fontname="Arial")

ax.set_xlabel('Time (Day)',fontsize=18, fontname="Arial")

ax.plot(Days, FGPT_hm/1000000, linewidth=3)

# inset axes....
axins = ax.inset_axes([0.5, 0.1, 0.4, 0.4])
axins.plot(Days, FGPT_hm/1000000)
# sub region of the original image
x1, x2, y1, y2 = 16741, 20637, 125, 145
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
#axins.set_xticklabels()
#axins.set_yticklabels()

ax.indicate_inset_zoom(axins, edgecolor="black")


plt.savefig("Realizatoins_Prediction_FGPT.jpg",dpi=100)
```
